 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

 Ans: To generate a dataset with categories and numerical values, and then visualize the distribution of a numerical variable across different categories, one common way to visualize this is by using a box plot or a violin plot. These plots allow you to see the distribution of numerical values within each category, including the spread, median, and potential outliers.

Here’s how you can generate a synthetic dataset and create a box plot to show the distribution of a numerical variable across different categories:

Example Code (Box Plot):

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a synthetic dataset
np.random.seed(42)

# Categories
categories = ['Category A', 'Category B', 'Category C', 'Category D']

# Numerical values for each category (randomized values for illustration)
data = {
    'Category A': np.random.normal(20, 5, 100),
    'Category B': np.random.normal(30, 7, 100),
    'Category C': np.random.normal(25, 4, 100),
    'Category D': np.random.normal(35, 6, 100)
}

# Create a DataFrame to organize the data for visualization
import pandas as pd

# Prepare data for plotting
df = pd.DataFrame({
    'Category': np.concatenate([np.repeat(cat, 100) for cat in categories]),
    'Value': np.concatenate([data[cat] for cat in categories])
})

# Create a box plot to visualize the distribution across categories
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Distribution of Values Across Categories')

# Show the plot
plt.show()
```

Explanation:
Dataset Generation:

np.random.normal(mean, std, size) is used to generate random numerical data for each category. The function generates random numbers drawn from a normal distribution.
In the example, we generate 100 values for each category, with different means and standard deviations to simulate varied distributions across categories.
Data Organization:

A Pandas DataFrame is created to combine the categories and corresponding numerical values into a structured format for easy visualization.
We concatenate the values for each category using np.concatenate and repeat the category names using np.repeat to match the numerical values.
Box Plot:

We use sns.boxplot() from the Seaborn library to create a box plot.
x='Category' specifies that categories will be plotted on the x-axis.
y='Value' specifies that the numerical values will be plotted on the y-axis.
palette='Set2' is used to apply a color palette to the boxes.
The box plot shows the median, interquartile range (IQR), and any potential outliers for each category.
Labels and Title:

plt.xlabel() and plt.ylabel() add labels for the x-axis and y-axis.
plt.title() adds a title to the plot.
Displaying the Plot:

plt.show() renders the plot.
Outcome:
The box plot will visualize the distribution of values for each category, allowing you to compare:

The median (central line within the box) for each category.
The spread (IQR, the box height) for each category.
The outliers (points outside the "whiskers" of the box) for each category.
This approach is particularly useful when you want to compare the spread and central tendency of numerical values across categories, as it highlights the variability, skewness, and potential outliers in the data.

Alternative: Violin Plot
If you want to visualize the distribution more smoothly, you can also use a violin plot, which combines aspects of a box plot and a kernel density plot to show the distribution's shape:

```python
# Create a violin plot to visualize the distribution across categories
plt.figure(figsize=(8, 6))
sns.violinplot(x='Category', y='Value', data=df, palette='Set2')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Violin Plot of Values Across Categories')

# Show the plot
plt.show()
```

The violin plot gives a more detailed representation of the data distribution, showing both the probability density and summary statistics (median, quartiles).

 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using heatmap.

 Ans: To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap, we can use the following steps:

Generate Synthetic Data: Create a dataset with correlated features. We can use numpy or pandas to generate data, and scipy.stats for introducing correlation between features.

Calculate the Correlation Matrix: We can use the .corr() method in Pandas or numpy.corrcoef() to calculate the correlation matrix between the features.

Visualize with a Heatmap: Use Seaborn’s heatmap() to visualize the correlation matrix, which provides a clean, intuitive way to show how features are correlated with each other.

Here’s how you can do it:

Example Code:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
n_samples = 500

# Create a random dataset with some correlation
# Feature 1 is random normal
feature_1 = np.random.normal(0, 1, n_samples)

# Feature 2 is correlated with Feature 1 (adding some noise)
feature_2 = feature_1 * 0.8 + np.random.normal(0, 0.5, n_samples)

# Feature 3 is also correlated with Feature 1, but weaker
feature_3 = feature_1 * 0.4 + np.random.normal(0, 0.5, n_samples)

# Feature 4 is random with weak correlation to others
feature_4 = np.random.normal(0, 1, n_samples)

# Create a DataFrame to organize the data
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Plot the heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5, fmt='.2f', cbar_kws={'shrink': 0.8})

# Add title and display the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Explanation:
Dataset Generation:

We create a dataset with 4 features (Feature 1, Feature 2, Feature 3, and Feature 4), where:
Feature 1 is generated using a standard normal distribution.
Feature 2 is linearly correlated with Feature 1 (with a correlation coefficient of 0.8).
Feature 3 is also correlated with Feature 1, but with a weaker correlation (0.4).
Feature 4 is generated independently (with no strong correlation).
Correlation Matrix Calculation:

corr_matrix = data.corr() calculates the Pearson correlation coefficients between each pair of features. The correlation values range from -1 (perfect negative correlation) to +1 (perfect positive correlation). A value of 0 indicates no correlation.
Heatmap Visualization:

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1) creates a heatmap of the correlation matrix.
annot=True adds the numerical correlation values on the heatmap cells.
cmap='coolwarm' chooses a color palette where red shades represent positive correlations and blue shades represent negative correlations.
vmin=-1 and vmax=1 set the color scale to the full range of correlation values.
fmt='.2f' formats the annotation to show two decimal places.
Display the Plot:

plt.title() adds a title to the heatmap.
plt.show() displays the plot.
Outcome:
The heatmap will display the correlation matrix of the dataset, where:

The diagonal elements will always be 1 because each feature is perfectly correlated with itself.
The off-diagonal elements show the strength of correlation between pairs of features. Strong positive correlations will be shown in red, strong negative correlations in blue, and weak or no correlation in light colors (e.g., white or light blue).
This visualization is useful for understanding how different features are related to one another in a dataset, and it helps identify features that are highly correlated, which might be useful for feature selection or multicollinearity detection.

Here are the steps involved:

Calculate the z values: Using the formula 
𝑧=sin(root x^2+Y^2)

Create the 3D surface plot: Using matplotlib to plot the 3D surface.
Here's the complete code to generate the 3D surface plot:

Code:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the x and y data using np.linspace
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create the meshgrid for x and y
x, y = np.meshgrid(x, y)

# Compute the z values based on the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays and create a DataFrame (optional)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a figure for the plot
fig = plt.figure(figsize=(10, 7))

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels for axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set title
ax.set_title(r'3D Surface Plot of $sin(\sqrt{x^2 + y^2})$')

# Show the plot
plt.show()
```

⁡, which generates a wave-like pattern.

Labels and Title: The ax.set_xlabel(), ax.set_ylabel(), and ax.set_zlabel() functions label the axes. The title is set with ax.set_title().

Customize the chart: We can customize the size of the bubbles and their transparency to make the chart more informative.
Here’s the full code to generate the bubble chart:

Code:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bubble chart
plt.figure(figsize=(10, 6))

# Bubble size is proportional to the Population (scaled for better visualization)
bubble_size = df['Population'] * 10  # Scaling factor for bubble size

# Plot the bubbles
plt.scatter(df['GDP'], df['Population'], s=bubble_size, alpha=0.6, c=df['Population'], cmap='viridis', edgecolors='w', linewidth=1)

# Add titles and labels
plt.title('Bubble Chart: Population vs GDP by Country', fontsize=14)
plt.xlabel('GDP (in Billion USD)', fontsize=12)
plt.ylabel('Population (in Millions)', fontsize=12)

# Annotate each bubble with the country's name
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i] + 10, df['Population'][i], country, fontsize=12, ha='left', va='bottom')

# Show the plot
plt.colorbar(label='Population Size (Color)')
plt.show()
```

edgecolors='w' and linewidth=1 add a white border around the bubbles for better visual clarity.

Each bubble is annotated with the corresponding country name, positioned slightly to the right of the bubble for clarity.
Colorbar:

A colorbar is added using plt.colorbar() to indicate that the colors of the bubbles represent the population size.

The x-axis represents the GDP.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource

# Set up random data
np.random.seed(42)

# Generate random data for x and y
x = np.random.rand(50) * 10  # 50 random x values between 0 and 10
y = np.random.rand(50) * 10  # 50 random y values between 0 and 10

# Generate random sizes and colors for the scatter plot
sizes = np.random.randint(10, 100, size=50)  # 50 random sizes between 10 and 100
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=50)  # 50 random colors

# Create a ColumnDataSource for Bokeh
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Set up the output file (this will save the plot as an HTML file)
output_file("scatter_plot.html")

# Create a figure object
p = figure(title="Scatter Plot with Random Sizes and Colors", x_axis_label='X', y_axis_label='Y', plot_height=400, plot_width=700)

# Add scatter plot using the data
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```

 5. Create a Bokeh heatmap using the provided dataset.

 Ans: To create a Bokeh heatmap using the provided dataset, we can follow these steps:

Steps:
Generate the Data:

data_heatmap: A 10x10 array of random values between 0 and 1.
x and y: Arrays of 10 values linearly spaced between 0 and 1.
xx, yy: A meshgrid of x and y values to represent the grid for the heatmap.
Set up the heatmap: Use p.rect() in Bokeh to create a heatmap by drawing rectangles whose colors correspond to the values in data_heatmap.

Display the plot: Customize and show the heatmap.

Code:

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.layouts import column

# Generate the data
data_heatmap = np.random.rand(10, 10)  # 10x10 matrix of random values between 0 and 1
x = np.linspace(0, 1, 10)  # 10 equally spaced points from 0 to 1 for x
y = np.linspace(0, 1, 10)  # 10 equally spaced points from 0 to 1 for y
xx, yy = np.meshgrid(x, y)  # Generate meshgrid for x and y

# Flatten the arrays to use as source for the plot
x_vals = xx.flatten()
y_vals = yy.flatten()
colors = data_heatmap.flatten()

# Set up the output file (this will save the plot as an HTML file)
output_file("heatmap.html")

# Create a figure object
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', plot_height=400, plot_width=400)

# Define a color mapper (Viridis palette)
color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

# Create the heatmap using rectangles
p.rect(x=x_vals, y=y_vals, width=0.1, height=0.1, source={'x': x_vals, 'y': y_vals, 'colors': colors},
       line_color=None, fill_color=color_mapper)

# Add a color bar to show the color scale
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```

Explanation:
Data:

data_heatmap = np.random.rand(10, 10) generates a 10x10 matrix of random values between 0 and 1.
x = np.linspace(0, 1, 10) and y = np.linspace(0, 1, 10) create arrays of 10 linearly spaced values between 0 and 1 for the x and y axes.
xx, yy = np.meshgrid(x, y) creates a meshgrid of x and y values, which represents the coordinates of the grid for the heatmap.
Flatten the Data:

We flatten the xx, yy, and data_heatmap arrays to make them one-dimensional arrays, which Bokeh can work with for plotting.
Bokeh Plot:

linear_cmap(): A function that creates a color mapper using a palette (in this case, Viridis256). It maps the colors array values (from data_heatmap) to corresponding colors.
p.rect(): This function draws rectangles on the plot. Each rectangle's position is defined by x_vals and y_vals, and the color of each rectangle is determined by the color_mapper.
The width=0.1 and height=0.1 set the size of each rectangle, effectively creating a 10x10 grid.
Color Bar:

The ColorBar is added to the right of the heatmap to show the color scale. The color scale is based on the values of data_heatmap.
Output:

output_file("heatmap.html") saves the plot as an HTML file.
show(p) displays the heatmap in the browser.
Expected Output:
The heatmap will be a 10x10 grid where each square's color corresponds to a value from the data_heatmap matrix.
The x-axis and y-axis represent the coordinates, ranging from 0 to 1.
The color bar on the right will show the range of values from the heatmap (from low to high) mapped to the colors.
Interactive Features:
The plot will be interactive, allowing you to zoom, pan, and hover over each cell to see the exact value.
